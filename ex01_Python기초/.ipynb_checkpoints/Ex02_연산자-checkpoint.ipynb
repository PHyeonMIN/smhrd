비교 연산자

```python
# 뭔가를 비교하게, 조건이 되어 질 수 있음
# >. ==. <, <=, >=, !=
a = 3
b = 7
print(a>b)
print(a<=b)
print(a==b)
print(a!=b)
```

```python
import random
a= random.randint(1,10)
b= random.randint(1,10)
print("a:{},b:{}".format(a,b))
print(a<b)
print(a>b)
print(a==b)
print(a!=b)
```

논리 연산자, 멤버 연산자

```python
# 논리 not, and, or
# 멤버 in, not in
```

```python
# not 논리 값을 뒤집는 역할
a=3
b=7
not a<b
not a==b
```

```python
# and, or 조건들을 연결시킴
not(a < b and a==b) # 둘다 True -> True
a < b or a==b
```

삼항 연산자  
a if 조건식 else b

```python
score = 80
'합격' if score>=60 else '불합격'
```

```python
a = int(input("정수입력 >> "))
b = int(input("정수입력 >> "))
'a : {}'.format(a) if a>b else 'b : {}'.format(b)
```

```python
a, b = map(int, input().split(' '))
print(a+b)
```
