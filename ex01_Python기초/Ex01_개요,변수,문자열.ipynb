#### 문자열 데이터 접근 
- 인덱싱 : 무엇인가를 하나하나 가리킴
- 슬라이싱 : 무엇인가를 잘라냄(범위)

문자열 인덱싱

```python
jy_info = 'My name is JY'
# M데이터에 접근 -> M 이  가지고 있는 인덱스 번호 불러야함
print(jy_info[0])
# i데이터에 접근 -> i가 위치한 인덱스 번호 불러야함
print(jy_info[8])
```

문자열 슬라이싱

```python
# 대상문자열[시작인덱스번호:끝 인덱스번호+1] 
# My 접근해보기
print(jy_info[:2])
```

```python
#name 데이터 접근
print(jy_info[3:7])
#JY 데이터 접근
print(jy_info[-2:])
#My name
print(jy_info[:7])
```

```python
jy_info
```

```python
# 슬라이싱 - 처음부터 끝까지 슬라이싱
# 문자열대상[시작값:끝값+1:증감값]
jy_info[0:14:2]
#문자열 뒤에서부터 앞으로 출력하기
jy_info[::-1]
```

문자열 관련 함수

```python
# 문자열에 포함된 문자 개수 세기
s = 'Hi, My name is JY'
s.count('i')
```

```python
# 문자열을 소문자 -> 대문자
s.upper().count('Y')
# 대문자 -> 소문자
s.lower().count('y')
```

```python
# find, index 문자 위치 알려주기
s.find('i')  # 문장안에서 첫번째로 만나는 i 인덱스 반환
s.find('z')  # 에러나지 않고 -1 반환
```

```python
s.index('i')
s.index('z') # ValueError: substring not found
```

```python
# rstrip()
s[:4].rstrip()
```

```python
# lstrip() 왼쪽공백제거
s[3:].lstrip()
```

```python
# 양 옆 공백 제거
s3 = '     빅데이터 화이팅!   '
s3.strip()
```

```python
# replace('문자1','문자2') : 문자열1을 문자열2로 바꿔줌

s3.replace('화이팅','파이팅!!!!')
```

```python
# split
s4 = '오늘도 고생하셨습니다, 조심히 들어가세요 ~'
s4.split(',')[1]   # list 구조
```
