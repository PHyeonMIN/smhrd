while

```python
# while 조건식 : 
    # 수행할 문장
```

```python
# 파이썬 최고 13번 출력해보기
num = 1
while num<=13:
    print(f'{num}. 파이썬 최고')
    num+=1
```

```python
# while 반복을 멈추게 하는 키워드 : break
num = 1 
while True:
    print('수행~~')
    if num ==2:
        break
    num+=1
```

```python
while True:
    num1 = int(input('첫번째 정수 입력>>>'))
    num2 = int(input('두번째 정수 입력>>>'))
    if num1==0 and num2==0:
        print('프로그램이 종료되었습니다.')
        break
    else:
        print('두 정수의 합 : {}'.format(num1+num2))
```

#### pickle
파이썬 자료형 그대로 반영해서 데이터를 저장하는 도구(모듈)

```python
import pickle
```

```python
with open('movie_review.pkl','rb') as f:
    data = pickle.load(f)
data[:5]
# [[리뷰문장,긍정1/부정0],[],....]
```

for문 이용해서 데이터 접근하는 것

```python
# 리뷰만 출력해보자
for r in data:
    print(r[0])
```

```python
# 긍정/부정 출력해보자
for r in data:
    print(r[1])
```

1. 전체 리뷰 개수 확인
: 리스트의 길이만 확인(len)

```python
len(data)
```

2. 리뷰 길이 평균 값

```python
total=0
for r in data:
    total += len(r[0])
total/len(data)
print(f'리뷰의 평균 길이는 \'{total/len(data)}\'입니다')
```

3. 긍정리뷰 / 부정리뷰 개수

```python
pos = 0
neg = 0
for r in data:
    if r[1]==1:
        pos+=1
    else:
        neg+=1
print(f"긍정 '{pos}'개")
print(f"부정 '{neg}'개")
```

4. 리뷰 길이 최대 값 / 리뷰 내용

```python
max_length =0
for r in data:
    if len(r[0]) > max_length:
        max_length = len(r[0])
        a = r[0]
print(f"리뷰 최대길이는 '{max_length}'입니다.")
print(a)
```

5. ㅋ가 포함된 리뷰 개수

```python
cnt = 0
for r in data:
    if 'ㅋ' in r[0]:
        cnt+=1
cnt
```

6. '검색 데이터'가 포함된 리뷰 개수

```python
cnt2=0
s = input('검색하시오 >> ')
for r in data:
    if s in r[0]:  #멤버연산자 -> '찾고싶은문자' in 찾을대상
        cnt2+=1
print(f"'{s}'가 포함된 리뷰는 {cnt2}개 입니다.")
```
