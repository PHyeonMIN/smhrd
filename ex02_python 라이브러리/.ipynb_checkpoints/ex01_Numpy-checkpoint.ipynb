### 사용하고자 하는 데이터 불러오기

```python
# loadtxt('파일명.확장자', delimiter='') : text 파일을 읽어 들일수 있는 함수
data = np.loadtxt('height_weight.txt',delimiter=',')
data
```

```python
# 전체 데이터로 부터 몸무게와 키의 데이터를 분리해 보자
# 키 ->height
# 몸무게 -> weight
height = data[0]
weight = data[1]
```

```python
height = height/100
height
```

```python
BMI = weight/(height**2)
BMI
```

```python
# sum() : 매개변수로 들어오는 모든 데이터들의 합을 계산하는 함수
arr = np.random.randint(1,10,size=(4,7))
arr
```

```python
print(arr.sum())
print(np.sum(arr))
```

```python
# mean() : 매개 변수의 모든 데이터에 대하여 평균을 구하는 함수
print(arr.mean())
print(np.mean(arr))
```

```python
# sqrt() : 제곱근을 구하는 함수
arr = np.arange(1,6)
print(arr)

print(np.sqrt(arr))
```

```python
# abs() : 절댓값을 구하는 함수
arr = np.array([-1,2,3,-4,5])
print(arr)

print(np.abs(arr))
```
