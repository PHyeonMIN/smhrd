### matplotlib란?
matplotlib은 파이썬에서 데이터를 차트나 플롯(Plot)으로 그려주는 시각화 라이브러리
라인 플롯, 바 차트, 파이차트, 히스토그램, Box Plot, Scatter Plot 등을 비롯한 다양한 차트와 플롯 스타일을 지원

### 라이브러리 사용

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3]
y=[2,4,6]
plt.plot(x,y)
plt.show()
```

#### line style

```python
y=[1,5,6]
plt.plot(y,ls='-.')
plt.show()
```

#### Line width
lw = 굵기(숫자로 지정)

```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]
plt.plot(x,y, lw='14', ls='--')
plt.show()
```

#### Marker

```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]
plt.plot(x,y,  lw='1', ls='--',marker='*')
plt.show()
```

### Line color

```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]
plt.plot(x,y,  lw='1', ls='--',marker='*',c='c')
plt.show()
```

#### Marker 스타일

```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]
plt.plot(x,y,  lw='1', ls='--',marker='o',c='c',ms=10, mfc='b')
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]
plt.plot(x,y,  lw='1', ls='--',marker='o',c='c',ms=10, mfc='y',mec="b",mew=5)
plt.show()
```

### 연습문제

```python
import numpy as np
x=np.arange(7)
y=[1,4,5,8,9,4,3]
plt.plot(x,y,ls='--',marker='o',color='g')
plt.show()
plt.plot(x,y,ls=':',marker='D',color='r',ms=10,mfc='b')
plt.show()
```

#### 범위 적용

```python
x=np.arange(7)
y=[1,4,5,8,9,4,3]
plt.plot(x,y,ls='--',marker='o',color='g')
plt.xlim(-2,10)
plt.ylim(-5,12)
plt.show()
```

#### 틱 설정
- 플롯이나 차트에서 축상의 위치 표시 지점을 틱(tick)이라고 함
- 틱에 써진 숫자 혹은 글자 틱 라벨(tick label)
- 틱의 위치나 틱 라벨은 Matplotlib가 자동으로 지정, 수동으로 설정하고 싶다면 xticks, yticks 명령을 이용

```python
x=np.arange(7)
y=[1,4,5,8,9,4,3]
plt.plot(x,y,ls=':',marker='D',color='r',mfc='b')
plt.xticks([0,3,6])
plt.yticks([1,5,9])
plt.show()
```

#### 그리드

```python
x=np.arange(7)
y=[1,4,5,8,9,4,3]
plt.plot(x,y,ls=':',marker='D',color='r',mfc='b')
plt.grid()
plt.show()
```

#### 여러개의 플롯

```python
x=[1,2,3,4]
y=[2,4,6,8]
z=[3,6,9]
plt.plot(x)
plt.plot(y)
plt.plot(z)
plt.grid()
plt.show()
```

#### 범례 표시

```python
x=[1,2,3,4]
y=[2,4,6,8]
z=[3,6,9]
plt.plot(x, label='박')
plt.plot(y, label='y')
plt.plot(z, label='z')
plt.legend() # 범례를 적용
plt.grid()
plt.show()
```

### <차대차 사건 중 죽거나 다친 사람이 많은 발생지 시도를 알아보고 시각화>

```python
data['사고유형_대분류'].value_counts()
```

```python
data.columns
```

```python
data1 = data[data['사고유형_대분류']=='차대차']
data1
```

```python
data1['사망사고']=data1['사망자수']+data1['사상자수']+data1['중상자수']+data1['경상자수']
data1
```

```python
data2 = data1[['사망사고','발생지시도']].groupby('발생지시도').sum()
data2
```

```python
x = data2.index
y = data2.iloc[:,0]
plt.bar(x,y)
plt.xlabel('발생지시도', labelpad=20)
plt.ylabel('사망사고')
plt.title('차대차 사건 중 죽거나 다친 사람이 많은 발생지 시도')
plt.show()
```
