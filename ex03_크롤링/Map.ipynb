### 데이터 프레임 (다이닝코드)

```python
driver = wb.Chrome()
url = 'https://www.diningcode.com/list.php?query=%EA%B4%91%EC%A3%BC%EA%B4%91%EC%97%AD%EC%8B%9C'
driver.get(url)

btn = driver.find_element_by_css_selector('#div_list_more')

for i in range(3) :
    btn.click()
    time.sleep(2)
    
soup = bs(driver.page_source, 'lxml')

res = soup.select('#div_list > li > a > span:nth-child(2)')
loc = soup.select('#div_list > li > a > span:nth-child(5)')
fav = soup.select('#div_list > li > p > span.favor.button')

delete = soup.select('#div_list > li > a > span:nth-child(5) > i')

for i in delete:
    i.extract()
    
res_list = []
loc_list = []
fav_list = []
num_list = []
    
for i in range(len(res)):
    res_list.append(res[i].text)
    loc_list.append(loc[i].text)

for i in range(len(res_list)):
    try :
        res_list[i] = res_list[i].split('.')[1].strip()
    except : 
        del res_list[i]
        
for i in range(len(loc_list)):
    try :
        loc_list[i] = loc_list[i].split('.')[1].strip()
    except : 
        del loc_list[i]
       
```

```python
res_list =[]
res_list2=[]
for i in range(len(res)):
    res_list.append(res[i].text)
for i in range(len(res_list)):
    try :
        res_list2.append(res_list[i].split('.')[1].strip())
    except : 
        continue
        
res_list2
```

```python
res_list[1]=res_list[1].split('.')[1].strip()
res_list
```

```python
dic = {'상호명' : res_list, '구' : loc_list, '좋아요' : fav_list, '번호' : num_list}
df = pd.DataFrame(dic)
df.set_index('번호', inplace = True)
pd.set_option('display.max_rows', None)
```

# 식신

```python
driver = wb.Chrome()
url = 'https://www.siksinhot.com/taste?upHpAreaId=5&hpAreaId=214&isBestOrd=N'
driver.get(url)
```

### 상호명 - 광산구 (식신)

```python
soup = bs(driver.page_source, 'lxml')
soup.select('#tabMove1 > div > ul > li > div > a > div > div > strong')
```

```python
res = soup.select('#tabMove1 > div > ul > li > div > a > div > div > strong')
```

### 즐겨찾기 - 광산구 (식신)

```python
soup = bs(driver.page_source, 'lxml')
soup.select('#tabMove1 > div > ul > li > div > a > div > ul.state_ul > li.ico_st02 > p')
```

```python
fav = soup.select('#tabMove1 > div > ul > li > div > a > div > ul.state_ul > li.ico_st02 > p')
```

### 데이터 프레임 (식신)

```python
res_list = []
loc_list = []
fav_list = []
num_list = []

for i in range(len(res)):
        res_list.append(res[i].text)
        loc_list.append('광산구')
        fav_list.append(fav[i].text)
        num_list.append(i+1)

dic = {'상호명' : res_list, '구' : loc_list, '즐겨찾기' : fav_list, '번호' : num_list}
df = pd.DataFrame(dic)
df.set_index('번호', inplace = True)
pd.set_option('display.max_rows', None)
```
