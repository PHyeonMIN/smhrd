### Q3

```python
driver = wb.Chrome()
url='http://corners.gmarket.co.kr/Bestsellers'
driver.get(url)
```

```python
title_list = []
price_list = []
rank_list = []
img = driver.find_elements_by_css_selector('img.lazy')

for i in range(20):
    img = driver.find_elements_by_css_selector('img.lazy')
    img[i].click()
    time.sleep(0.1)
    
    soup = bs(driver.page_source,'lxml')
    title = soup.select_one('h1.itemtit')    # 상품명이 하나라서 select_one
    price = soup.select_one('strong.price_real')
    title_list.append(title.text)
    price_list.append(price.text)
    rank_list.append(i+1)
    
    driver.back()
    time.sleep(0.1)
```

```python
print(len(title_list))
print(len(price_list))
print(len(rank_list))
```

```python
dic ={'상품명':title_list, '가격':price_list, '순위':rank_list}
```

```python
Gmarket = pd.DataFrame(dic)
Gmarket.set_index('순위',inplace=True)
```

```python
Gmarket
```

### Q4

```python
url = 'https://www.melon.com/chart/index.htm'
h={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.45 Safari/537.36'}
res = req.get(url, headers = h)
soup = bs(res.text, 'lxml')
title = soup.select('div.ellipsis.rank01>span>a')
singer = soup.select('div.ellipsis.rank02>a:nth-child(1)')
```

```python
print(len(title))
print(len(singer))
```

```python
title_list = []
singer_list = []
rank_list=[]
for i in range(len(title)):
    title_list.append(title[i].text)
    singer_list.append(singer[i].text)
    rank_list.append(i+1)
```

```python
dic = {'곡명':title_list, '가수명':singer_list, '순위':rank_list}
melon = pd.DataFrame(dic)
melon.set_index('순위', inplace=True)
```

```python
melon
```

```python
melon.to_csv('./melon.csv', encoding='utf-8')
```

### Q5

```python
driver = wb.Chrome()
url='https://www.youtube.com/'
driver.get(url)
time.sleep(0.5)
```

```python
input_area = driver.find_element_by_css_selector('input#search')
input_area.send_keys('아이유')
input_area.send_keys(Keys.ENTER)
time.sleep(0.5)
```

```python
body = driver.find_element_by_css_selector('body')
for i in range(50):
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.5) # 0.5초
```

```python
soup = bs(driver.page_source,'lxml')
title = soup.select('a#video-title')
lookup_cnt = soup.select('#metadata-line > span:nth-child(1)')
```

```python
for i in lookup_cnt:
    print(i)
```

```python
print(len(title))
print(len(lookup_cnt))
```

```python
for i in range(len(title)):
    print(title[i].text)
```

```python
for i in range(0,len(lookup_cnt)):
    print(lookup_cnt[i].text)
```

```python
for i in range(33,len(lookup_cnt)):
    print(lookup_cnt[i].text)
# 0~32까지 데이터를 어디서 가지고 오는지 모르겠습니다.
# 제목과 일치하는 조회수는 33번부터임 무조건 그래서 데이터 전처리 과정을 추가했습니다.
# 또한 플레이리스트는 조회수가 아닌 업데이트: 오늘 라고 표시됨 데이터 전처리할것
```

```python
title_list =[]
lookup_cnt_list = []
cnt_list=[]
for i in  range(len(title)):
    title_list.append(title[i].text.replace('\n','')) 
    # \n안들어가게 태그를 잡으면 안들어가는 제목이있음 그래서 replace대체
    cnt_list.append(i+1)
    
for i in range(33,len(lookup_cnt)):
    if(lookup_cnt[i].text=='업데이트: 오늘'):
        print('삭제')
    else:
        lookup_cnt_list.append(lookup_cnt[i].text)
    # 제목과 일치하는 조회수는 33번부터이기떄문에 33번부터 리스트 저장
```

```python
print(len(title_list))
print(len(lookup_cnt_list))
print(len(cnt_list))
```

```python
dic = {'제목':title_list, '조회수':lookup_cnt_list, '순번':cnt_list}
youtube = pd.DataFrame(dic)
youtube.set_index('순번', inplace=True)
```

```python
youtube
```
