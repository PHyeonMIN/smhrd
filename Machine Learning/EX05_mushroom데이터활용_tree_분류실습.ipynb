```python
# 도구
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier # 분류 예측기
```

- poisonous 컬럼 확인 : 답(데이터 표현 확인)

```python
# 컬럼 인덱싱
data['poisonous']
# data.iloc[:,0]
# data.loc[:,'poisonous']

# 클래스의 개수 확인
data['poisonous'].value_counts()
# e 식용 - 4208, p 독버섯 - 3916
```

### 문제(독립변수, 속성, 특성, 피쳐)와 답(label, 종속변수, target) 분리

```python
X = data.iloc[:,1:]
y= data['poisonous']
```

추가설명) 라벨인코딩(레이블 인코딩)

```python
# 서식지 컬럼 활용하여 범주 -> 수치화
X['habitat'].unique() # 서식지의 종류 출력

# 레이블인코딩을 위한 딕셔너리 정의
hab_dict = {'u':2, 'g':1, 'm':3, 'd':5, 'p':4, 'w':6, 'l':7}

# 서식지 컬럼과 hab_dict 연결
# map : 값을 맵핑 ( 같은 키값을 찾으면 벨류 값으로 변경 )
X['habitat'].map(hab_dict)
```

#### 원핫인코딩(one-hot encoding)
- 값의 크고 작음의 의미가 없을때 주로 사용

```python
X_one_hot = pd.get_dummies(X)
X_one_hot.head()
```

### train data, test data로 분리

```python
X_train, X_test, y_train, y_test = train_test_split(X_one_hot,y, test_size= 0.3, random_state=7)
```

```python
# 데이터의 크기들 확인해보기
print('훈련용 문제 : ',X_train.shape)
print('훈련용 답 : ',X_test.shape)
print('테스트용 문제 : ',y_train.shape)
print('테스트용 답 : ',y_test.shape)
```

### 모델링

```python
# 모델 객체 생성
tree_model = DecisionTreeClassifier() # 가지칠 수 있을만큼 가지를 뻗어나가게 됨 -> 과대적합
```

```python
# 모델 학습 진행
tree_model.fit(X_train,y_train)
```

```python
# 예측 predict(생략)
pre = tree_model.predict(X_test)
```

```python
# 모델 성능 평가
# test용 데이터를 활용
# 모델 자체 score 함수 활용
tree_model.score(X_test,y_test)
```

```python
# 우리가 보유한 데이터가 한정적임
# 여러번의 검증은 어떻게 실시할 수 있는가?
# 보유한 데이터를 활용해서 이안에서 여러번 train,test를 분리해서 학습하고 성능 평가해야함
# 교차검증 cross_validation
```

### 교차검증

```python
from sklearn.model_selection import cross_val_score 
# 과대적합제어 자체x / 방지를 해줄 수 있음
# 일반화 성능 정도를 파악
# 교차검증은 객체 생성 후, fit,score 뽑기 전에도 실행
```

```python
tree_result = cross_val_score(tree_model,X_train,y_train, cv=5)
tree_result

# 현재 tree_model 일반화 되었음을 알 수 있음
# 따로 과대적합 제어하지 않아도 상관없음
print('교차검증 정확도 평균 : ', tree_result.mean())
```

#### tree 학습 현황 시각화

```python
from sklearn.tree import export_graphviz
export_graphviz(tree_model, out_file='tree.dot',
               class_names=['독','식용'],
               feature_names=X_one_hot.columns,
               impurity=True,
               filled=True)
```

```python
# 그래프 출력
import graphviz

# 파일 입출력 사용하는 코드(파일 불러오는 , 출력하는 코드)
with open('tree.dot', encoding='UTF8') as f:
    dot_graph = f.read()

display(graphviz.Source(dot_graph))
```

#### 과대적합 제어 하이퍼 파라미터 설정

- max_depth 트리 깊이 설정

```python
# 객체 생성
tree_model2 = DecisionTreeClassifier(max_depth = 3)
# 학습
tree_model2.fit(X_train,y_train)
# 교차검증 성능 확인하기
# 교차검증 횟수 == 나누는 횟수
tree_result2 = cross_val_score(tree_model2,X_train,y_train,cv=5)
print(tree_result2)
print(tree_result2.mean())
```

- 그래프화

```python
from sklearn.tree import export_graphviz
export_graphviz(tree_model2, out_file='tree.dot',
               class_names=['독','식용'],
               feature_names=X_one_hot.columns,
               impurity=True,
               filled=True)
```

```python
# 그래프 출력
import graphviz

# 파일 입출력 사용하는 코드(파일 불러오는 , 출력하는 코드)
with open('tree.dot', encoding='UTF8') as f:
    dot_graph = f.read()

display(graphviz.Source(dot_graph))
```
